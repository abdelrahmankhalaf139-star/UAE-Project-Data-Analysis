```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
df=pd.read_excel("/content/UAE_ProjectListing (1).xlsx")
df.head()
```

```python
plt.figure(figsize=(10,5));sns.countplot(data=df,x="City");plt.xticks(rotation=30);plt.show()
plt.figure(figsize=(10,5));sns.countplot(data=df,x="Industry");plt.xticks(rotation=30);plt.show()
plt.figure(figsize=(10,5));sns.scatterplot(data=df,x="Completion Year",y="Project Value ($m)",hue="Industry");plt.show()
```

```python
print(df.groupby("Industry")["Project Value ($m)"].mean())
vals=df["Project Value ($m)"]
print(vals.mean(),vals.min(),np.percentile(vals,25),np.percentile(vals,50),np.percentile(vals,75),vals.max())
```

```python
plt.figure(figsize=(10,6));sns.boxplot(x=df["Completion Year"]);plt.show()
plt.figure(figsize=(10,5));sns.histplot(df["Project Value ($m)"],bins=20,edgecolor="black");plt.show()
```

```python
construction=df[df["Industry"]=="Construction"]
plt.figure(figsize=(8,5));sns.countplot(data=construction,x="Sector");plt.xticks(rotation=30);plt.show()
power=df[df["Industry"]=="Power"]
plt.figure(figsize=(8,5));sns.countplot(data=power,x="Project Status");plt.xticks(rotation=30);plt.show()
```

```python
relation=df[["Project Value ($m)","Completion Year"]]
sns.heatmap(relation.corr(),annot=True);plt.show()
print(relation.corr().loc["Project Value ($m)","Completion Year"])
```
